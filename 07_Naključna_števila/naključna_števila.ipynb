```python
import matplotlib.pyplot as plt
```

$$X_{n+1} = (a * X_{n} + c) \mod m $$

```python
def LCG(X0:int,a:int,c:int,m:int):
    '''
    X0: seme
    a: mno≈æitelj (angl. multiplier),
    c: inkrement (angl. increment),
    m: modulo (angl. modulus).
    '''
    return (a*X0+c)%m

def generator(X0,a,c,m,N=1):
    if N==1:
        return LCG(X0,a,c,m)/m
    else:
        sample=[]
        sample.append(X0/m)
        for k in range(N-1):
            X1=LCG(X0,a,c,m)
            X0=X1
            sample.append(X0/m)
        return sample
    
def plot(sample):
    plt.figure(figsize=(21,6))
    plt.subplot(1,3,1)
    plt.plot(sample,'o-')
    plt.xlabel('$n$')
    plt.ylabel('$X_n$')
    
    plt.subplot(1,3,2)
    x=sample[0:-1]
    y=sample[1:]
    plt.scatter(x,y)
    plt.xlabel('$X_n$')
    plt.ylabel('$X_{n+1}$')
    
    plt.subplot(1,3,3)
    plt.hist(sample,bins=20)
    
    plt.show()
```

# Primer 1

```python
a=c=x0=5
m=10

sample=generator(x0,a,c,m,N=10)
print("Sample =",sample)
plot(sample)
```

# Primer 2

```python
a=19
c=0
x0=5
m=256

sample=generator(x0,a,c,m,N=100)
print("Sample =",sample)
plot(sample)

sample1=generator(x0,a,c,m,N=100)
sample2=generator(x0+1,a,c,m,N=100)
sample3=generator(x0+2,a,c,m,N=100)

plt.title("Vpliv semena na potek vrste")
plt.plot(sample1,'o-',label='sample1')
plt.plot(sample2,'o-',label='sample2')
plt.plot(sample3,'o-',label='sample3')
plt.xlim(0,20)
plt.show()
```
